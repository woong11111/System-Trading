## 1. All Coin at Bithumb

```python
import pybithumb
import numpy as np

def get_hpr(ticker, period) :
    try :
        df = pybithumb.get_ohlcv(ticker)
        df = df[period]
        
        df['ma5'] = df['close'].rolling(window=5).mean().shift(1)
        df['range'] = (df['high'] - df['low']) * 0.5
        df['target'] = df['open'] + df['range'].shift(1)
        df['bull'] = df['open'] > df['ma5']
        
        fee = 0.005
        df['ror'] = np.where((df['high'] > df['target']) & df['bull'], df['close'] / df['target'] - fee, 1)
        
        df['hpr'] = df['ror'].cumprod()
        
        return df['hpr'][-2]
    
    except : 
        return 1
    

tickers = pybithumb.get_tickers()
periods = ['2019-01', '2019-02', '2019-03', '2019-04', '2019-05', '2019-06']

hprs = []

for period in periods :
    for ticker in tickers :
        hpr = get_hpr(ticker, period)
        hprs.append((ticker, hpr, period))
        
sorted_hprs = sorted(hprs, key = lambda x : x[1], reverse=True)
print(sorted_hprs[:70])
```

```python
import pandas as pd
df = pd.DataFrame(sorted_hprs, columns = ['Ticker', 'Holding Period Ratio', 'Period'])
```

```python
df_2019_01 = df[df['Period'] == '2019-01'][:5]
df_2019_02 = df[df['Period'] == '2019-02'][:5]
df_2019_03 = df[df['Period'] == '2019-03'][:5]
df_2019_04 = df[df['Period'] == '2019-04'][:5]
df_2019_05 = df[df['Period'] == '2019-05'][:5]
df_2019_06 = df[df['Period'] == '2019-06'][:5]
```

```python
pd.concat([df_2019_01, df_2019_02, df_2019_03, df_2019_04, df_2019_05, df_2019_06], axis=0)
```
