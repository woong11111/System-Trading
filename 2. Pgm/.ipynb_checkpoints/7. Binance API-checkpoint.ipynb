```python
import ccxt
```

```python
binance = ccxt.binance()
```

```python
ohlcvs = binance.fetch_ohlcv("BTC/USDT",  timeframe='1d')

import pandas as pd

df = pd.DataFrame(ohlcvs, columns = ['일자', '시가', '고가', '저가', '종가', '거래량'])
df['일자'] = pd.to_datetime(df['일자']/1000, unit = 's')
```
