## 1. OHLCV 활용한 백테스팅 구조

```python
import pybithumb

df = pybithumb.get_ohlcv("BTC")
```

```python
df.tail()
```

```python
df['range'] = (df['high'] - df['low']) * 0.5
```

```python
df.tail()
```

```python
df['range_shift1'] = df['range'].shift(1)
df.tail()
```

```python
df['target'] = df['open'] + df['range_shift1']
df.tail()
```

```python
import numpy as np

fee = 0.002
df['ror'] = np.where(df['high'] > df['target'], df['close'] / df['target'] - fee, 1)  #ror = rate of return
# np.where(조건, 조건이 참일 때의 값, 조건이 거짓일 때의 값)

df.tail(20)
```

## 3. 최적 Range 산출

```python
def get_ror(k, ticker) :
    df = pybithumb.get_ohlcv(ticker)
    df = df.iloc[-30:] # 30일 구간 ror 산출
    df['range'] = (df['high'] - df['low']) * k
    df['target'] = df['open'] + df['range'].shift(1)
    
    fee = 0.002
    df['ror'] = np.where(df['high'] > df['target'], df['close'] / df['target'] - fee, 1)
    
    ror = df['ror'].cumprod()[-2]
    return ror

ticker_result = []
k_result = []
ror_result = []

#tickers = pybithumb.get_tickers()
tickers = ['BTC', "ZIL", "ETH", "BCH"]
for ticker in tickers :
    for k in np.arange(0.1, 1.0, 0.1):
        ror = get_ror(k, ticker)
        ticker_result.append(ticker)
        k_result.append(k)
        ror_result.append(ror)
        #print(ticker, '-'*50)
        #print("%.1f %.3f" %(k, ror))
```
